```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4" 
model = hub.load(module_url)
print ("module %s loaded" % module_url)
with open('STS2016.input.postediting.txt') as f:
    lines = str(f.readlines())
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

s1=str(lines).split('\\n')
line1=[]
line2=[]
for i in range(0,len(s1)-1):
  line1.append(s1[i].split('\\t')[0])
  line2.append(s1[i].split('\\t')[1])
import re
string1=[]
for i in line1:
  string1.append(re.sub(r"[^a-zA-Z0-9]+",' ',i))
string2=[]
for i in line2:
  string2.append(re.sub(r"[^a-zA-Z0-9]+",' ',i))
strin=string1 +string2
sentence_embeddings = model(strin)
import numpy as np
x=0
similar=[]
for i in range(0, len(string1)-1):
  query = string1[i]
  query_vec = model([query])[0]
  sim = cosine(query_vec, model([string2[i]])[0])
  print("Sentence = ", string1[i], string2[i], "; similarity = ", sim)
  similar.append(sim)
  x=x+1
```

```python
final=[]
for i in similar:
  final.append(i*5)
print(final)
```
